```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Open webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convert frame to RGB (MediaPipe requires RGB format)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_frame)
    
    # Draw landmarks if hands are detected
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
    
    # Display the frame
    cv2.imshow("Hand Tracking", frame)
    
    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import json
import os
from collections import defaultdict

# Paths
json_path = r"D:\FYPSHIT\DATASET\WLASL_v0.3.json"  # Update this
video_dir = r"D:\FYPSHIT\DATASET\videos"  # Update this
missing_videos_txt = r"D:\FYPSHIT\DATASET\missing.txt"  # TXT file containing missing video IDs

# Step 1: Load Missing Video IDs from the TXT File
missing_videos = set()

with open(missing_videos_txt, "r") as f:
    for line in f:
        missing_videos.add(line.strip())  # Remove newline characters

print(f"✅ Loaded {len(missing_videos)} missing video IDs.")

# Step 2: Load the Original JSON File
with open(json_path, "r") as f:
    data = json.load(f)

# Step 3: Filter out missing videos and count available videos per class
cleaned_data = []
class_counts = defaultdict(int)  # Dictionary to store video count per class

for word in data:
    gloss = word["gloss"]  # ASL word label
    valid_instances = []

    for instance in word["instances"]:
        video_id = instance["video_id"]
        video_path = os.path.join(video_dir, f"{video_id}.mp4")

        # Check if the video exists and is not in the missing list
        if video_id not in missing_videos and os.path.exists(video_path):
            valid_instances.append(instance)

    # Only keep classes that have at least one valid video
    if valid_instances:
        class_counts[gloss] = len(valid_instances)
        cleaned_data.append({"gloss": gloss, "instances": valid_instances})

# Step 4: Save the Cleaned JSON Data
cleaned_json_path = r"D:\FYPSHIT\DATASET\WLASL_cleaned.json"  # Update if needed

with open(cleaned_json_path, "w") as f:
    json.dump(cleaned_data, f, indent=4)

# Step 5: Save Class Counts to a Separate File
class_counts_path = r"D:\FYPSHIT\DATASET\class_counts.json"  # Update if needed

with open(class_counts_path, "w") as f:
    json.dump(class_counts, f, indent=4)

# Print Summary
print(f"✅ Cleaned JSON saved to: {cleaned_json_path}")
print(f"✅ Class counts saved to: {class_counts_path}")
print(f"📊 Total words (classes) with valid videos: {len(class_counts)}")
```

```python
import json
import os
from collections import defaultdict

# Paths (Update if needed)
json_path = r"D:\FYPSHIT\DATASET\WLASL_v0.3.json"  # Original JSON
video_dir = r"D:\FYPSHIT\DATASET\videos"  # Video folder

# Load the original JSON file
with open(json_path, "r") as f:
    data = json.load(f)

# Step 1: Count total videos in the directory
video_files = set(os.listdir(video_dir))  # Get all filenames in the folder
total_videos_in_dir = len(video_files)

print(f"📂 Total videos available in directory: {total_videos_in_dir}")

# Step 2: Count videos for each class in the original JSON
class_video_counts = defaultdict(int)

for word in data:
    gloss = word["gloss"]  # ASL word label
    for instance in word["instances"]:
        video_id = instance["video_id"]
        video_filename = f"{video_id}.mp4"

        if video_filename in video_files:
            class_video_counts[gloss] += 1

# Step 3: Print results
print("\n📊 Videos per class (word label):")
for gloss, count in sorted(class_video_counts.items(), key=lambda x: -x[1]):
    print(f"{gloss}: {count} videos")

# Step 4: Save the class-wise video counts to a JSON file
output_class_counts = r"D:\FYPSHIT\DATASET\original_class_counts.json"  # Update if needed
with open(output_class_counts, "w") as f:
    json.dump(class_video_counts, f, indent=4)

print(f"\n✅ Class-wise video counts saved to: {output_class_counts}")
```
